```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Load Dataset
file_path = "/content/titanic (2).csv"  # Use correct file path
df = pd.read_csv(file_path)
```

```python
# 2. Data Overview
print("\nDataset Shape:", df.shape)
print("\nFirst 5 Rows:\n", df.head())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# 3. Handling Missing Values
df.fillna(df.median(numeric_only=True), inplace=True)  # Fill numeric missing values with median
df.fillna(df.mode().iloc[0], inplace=True)  # Fill categorical missing values with mode
```

```python
# 4. Correlation Analysis (Only Numeric Columns)
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['number'])  # Select numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 5. Visualizing Trends, Patterns, and Outliers
```

```python
## Survival Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(x="Survived", data=df, palette="coolwarm")
plt.title("Survival Distribution")
plt.show()
```

```python
## Boxplot for Outlier Detection
plt.figure(figsize=(10, 5))
sns.boxplot(data=numeric_df)
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

```python
## Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=30, kde=True, color="blue")
plt.title("Age Distribution")
plt.show()
```

```python
## Fare vs. Survival
plt.figure(figsize=(8, 5))
sns.boxplot(x="Survived", y="Fare", data=df, palette="coolwarm")
plt.title("Fare vs. Survival")
plt.show()
```

```python
## Passenger Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x="Pclass", data=df, hue="Survived", palette="coolwarm")
plt.title("Passenger Class vs. Survival")
plt.show()
```
