```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "/content/titanic.csv"
df = pd.read_csv(file_path)
df
```

```python
print("\nNull Values:")
print(df.isnull().sum())
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64'])
```

```python
print("\nStatistics:")
print("Mean:")
print(numerical_cols.mean())
```

```python
normalized_data = (numerical_cols - numerical_cols.min()) / (numerical_cols.max() - numerical_cols.min())
print("\nNormalized Data (First 5 rows):")
print(normalized_data.head())
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Age'], df['Fare'], 'bo', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
df_sorted = df.sort_values(by='Age')
df_sorted['Cumulative Survival'] = df_sorted['Survived'].cumsum()
plt.plot(df_sorted['Age'], df_sorted['Cumulative Survival'])
plt.xlabel('Age')
plt.ylabel('Cumulative Survival')
plt.title('Cumulative Survival by Age')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Survival Ratio')
plt.show()
```
